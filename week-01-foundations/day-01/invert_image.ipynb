```python
# Step 1: Import necessary libraries
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Load the image
image_path = 'avatar.jpeg'
img = Image.open(image_path)
img = img.convert('RGB')  # ensure image is in RGB

# Step 3: Convert image to NumPy array
img_array = np.array(img)
print("Original image shape:", img_array.shape)

# Step 4: Invert the colors
inverted_array = 255 - img_array  # simple inversion: 255 - pixel value

# Step 5: Convert back to image
inverted_img = Image.fromarray(inverted_array)

# Step 6: Save the inverted image
inverted_img.save('avatar_inverted.jpeg')

# Step 7: Display original and inverted images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(inverted_img)
plt.title('Inverted Image')
plt.axis('off')

plt.show()
```
